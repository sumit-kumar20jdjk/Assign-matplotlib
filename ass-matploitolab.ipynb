```python
'''Q-2ans- A scatter plot is a type of data visualization that displays individual data points on a two-dimensional 
   graph. Each point on the graph represents the values of two variables, one plotted along the x-axis and the other 
   along the y-axis. Scatter plots are particularly useful for visualizing the relationship between two continuous 
    variables and identifying patterns, trends, or correlations within the data.'''
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
plt.show()

```

```python
'''Q-3 ans-The subplot() function in Matplotlib is used to create a grid of subplots within a single figure.This
function is particularly helpful when you want to display multiple plots or charts in a single figure, making it 
easier to compare different aspects of your data. The subplot() function takes three integer arguments: the number 
of rows, the number of columns, and the index of the subplot being created.'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])


plt.subplot(2,2, 1)  
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)  
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)  
plt.plot(x3, y3)
plt.title('Line 3')


plt.subplot(2, 2, 4)  
plt.plot(x4, y4)
plt.title('Line 4')
plt.tight_layout()
plt.show()

```

```python
'''Q4ans- A bar plot (or bar chart) is a data visualization that presents categorical data with rectangular bars. The length
  of each bar is proportional to the quantity it represents. Bar plots are commonly used to compare and display the values of 
  different categories or groups, making them particularly effective for summarizing and visualizing the distribution of data.
  Comparison of Categories: Bar plots are effective for comparing the values of different categories or groups. They provide
  a clear visual representation of how quantities differ across distinct entities.

Categorical Data Distribution: Bar plots are useful for visualizing the distribution of categorical data, making it easy to 
identify patterns, trends, and disparities among different groups.

Frequency Distribution: In the context of histograms, which are a type of bar plot, they are used to display the frequency
distribution of continuous data when grouped into discrete bins.

Part-to-Whole Relationships: Bar plots can be used to illustrate the part-to-whole relationships within a single category,
such as the contribution of each subcategory to the total.

Comparison over Time: Time-series bar plots can be created to visualize changes in categorical data over time, allowing for 
trend analysis.'''
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel('company')
plt.ylabel('profit')
plt.show()
```

```python
'''Q5-ans-A box plot, also known as a box-and-whisker plot, is a statistical data visualization that provides a summary 
of the distribution of a dataset. It displays the distribution of a dataset based on a five-number summary: minimum, first 
quartile (Q1), median, third quartile (Q3), and maximum. The box plot is particularly useful for understanding the central 
tendency, spread, and presence of outliers in a dataset.

Summary of Distribution: Box plots provide a concise and visual summary of the distribution of a dataset, offering insights
into its central tendency, variability, and skewness.

Identification of Outliers: Box plots are effective in identifying outliers in a dataset. Outliers are often considered data 
points that fall outside the whiskers of the box plot.

Comparison of Distributions: Box plots are useful for comparing the distributions of different datasets or groups. Multiple
box plots can be displayed side by side for easy comparison.

Understanding Skewness: The asymmetry or skewness of a distribution can be visually assessed by observing the relative lengths
of the whiskers.

Visualization of Quartiles: Quartiles (Q1, Q2, Q3) and the interquartile range (IQR) are visually represented, aiding in a 
deeper understanding of the dataset's spread.'''
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


plt.boxplot([box1, box2])


plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plots of Two Distributions')


plt.show()
```
